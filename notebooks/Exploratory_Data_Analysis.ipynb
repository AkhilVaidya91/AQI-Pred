__Outlier Handline Method__: To handle the lower outliers, the values lower than Q1 - 1.5 x IQR were replaced with the lower boundary value for the same and to handle the higher outliers, values higher than Q3 + 1.5 x IQR were replaced with the upper boundary. Here, Q1 is 25th quartile, Q2 is 50th quartile, Q3 is 75th quartile.

```python
## Outlier handling

Q1 = df_numeric.quantile(0.25)
Q3 = df_numeric.quantile(0.75)

IQR = Q3 - Q1

upper_limit = Q3 + 1.5 * IQR
lower_limit = Q1 - 1.5 * IQR
```

```python
upper_limit.iloc[0]
```

```python
## replacing the outliers with the upper and lower limits

for i in df_numeric.columns:
    df_numeric[i] = np.where(df_numeric[i] > upper_limit[i], upper_limit[i], df_numeric[i])
    df_numeric[i] = np.where(df_numeric[i] < lower_limit[i], lower_limit[i], df_numeric[i])
```

```python
def calculate_aqi(row):
    """
    Calculate AQI for each pollutant and return the maximum AQI.
    I = ((Ihigh - Ilow)/(Chigh - Clow)) * (C - Clow) + Ilow
    """
    # Adjusted AQI breakpoint tables with a larger reduction in Chigh for each pollutant
    aqi_breakpoints = {
        'PM2.5': [
            (0.0, 3.0, 0, 50),
            (3.1, 8.85, 51, 100),
            (8.86, 13.85, 101, 150),
            (13.86, 37.6, 151, 200),
            (37.7, 62.6, 201, 300),
            (62.7, 87.6, 301, 400),
            (87.7, 125.2, 401, 500)
        ],
        'PM10': [
            (0, 13.5, 0, 50),
            (13.6, 38.5, 51, 100),
            (38.6, 63.5, 101, 150),
            (63.6, 88.5, 151, 200),
            (88.6, 106.0, 201, 300),
            (106.1, 126.0, 301, 400),
            (126.1, 151.0, 401, 500)
        ],
        'SO2': [
            (0, 8.75, 0, 50),
            (8.76, 18.75, 51, 100),
            (18.76, 46.25, 101, 150),
            (46.26, 76.0, 151, 200),
            (76.1, 151.0, 201, 300),
            (151.1, 201.0, 301, 400),
            (201.1, 251.0, 401, 500)
        ],
        'NO2': [
            (0, 1.55, 0, 6.25),
            (1.56, 3.1, 6.25, 12.5),
            (3.15, 11.25, 12.5, 17.5),
            (11.26, 18.75, 17.5, 25),
            (18.76, 37.5, 25, 37.5),
            (37.51, 50.0, 37.5, 50),
            (50.1, 62.5, 50, 62.5)
        ],
        'CO': [
            (0.0, 1.1, 0, 50),
            (1.2, 2.35, 51, 100),
            (2.4, 3.1, 101, 150),
            (3.2, 3.85, 151, 200),
            (3.9, 7.6, 201, 300),
            (7.7, 10.1, 301, 400),
            (10.2, 12.6, 401, 500)
        ],
        'O3': [
            (0, 13.5, 0, 50),
            (13.6, 17.5, 51, 100),
            (17.6, 21.25, 101, 150),
            (21.26, 26.25, 151, 200),
            (26.26, 50.0, 201, 300),
            (50.1, 126.0, 301, 400),
            (126.1, 151.0, 401, 500)
        ]
    }




    
    def calculate_pollutant_aqi(concentration, breakpoints):
        if pd.isna(concentration):
            return np.nan
            
        for Clow, Chigh, Ilow, Ihigh in breakpoints:
            if Clow <= concentration <= Chigh:
                return ((Ihigh - Ilow) / (Chigh - Clow)) * (concentration - Clow) + Ilow
        return np.nan

    # Calculate AQI for each pollutant using max values
    aqi_values = []
    
    # PM2.5
    if not pd.isna(row['PM2_MAX']):
        pm25_aqi = calculate_pollutant_aqi(row['PM2_MAX'], aqi_breakpoints['PM2.5'])
        aqi_values.append(pm25_aqi)
    
    # PM10
    if not pd.isna(row['PM10_MAX']):
        pm10_aqi = calculate_pollutant_aqi(row['PM10_MAX'], aqi_breakpoints['PM10'])
        aqi_values.append(pm10_aqi)
    
    # SO2
    if not pd.isna(row['SO2_MAX']):
        so2_aqi = calculate_pollutant_aqi(row['SO2_MAX'], aqi_breakpoints['SO2'])
        aqi_values.append(so2_aqi)
    
    # NO2
    if not pd.isna(row['NO2_MAX']):
        no2_aqi = calculate_pollutant_aqi(row['NO2_MAX'], aqi_breakpoints['NO2'])
        aqi_values.append(no2_aqi)
    
    # CO
    if not pd.isna(row['CO_MAX']):
        co_aqi = calculate_pollutant_aqi(row['CO_MAX'], aqi_breakpoints['CO'])
        aqi_values.append(co_aqi)
    
    # Ozone
    if not pd.isna(row['OZONE_MAX']):
        o3_aqi = calculate_pollutant_aqi(row['OZONE_MAX'], aqi_breakpoints['O3'])
        aqi_values.append(o3_aqi)
    
    # Return the maximum AQI value among all pollutants
    if aqi_values:
        return max(filter(lambda x: not pd.isna(x), aqi_values))
    return np.nan

def calculate_aqi_for_dataframe(df):
    return df.apply(calculate_aqi, axis=1)
```

```python
AQI_values = calculate_aqi_for_dataframe(df_cleaned)
AQI_values
```

```python
## fix the distribution

def fix(df):
    # Check if 'AQI' column exists in the DataFrame
    if 'AQI' in df.columns:
        # Apply conditions using a lambda function
        df['AQI'] = df['AQI'].apply(lambda x: x - 40 if 400 <= x <= 500 else 
                                    x - 30 if 300 <= x <= 400 else
                                    x - 50 if 200 <= x <= 300 else x)
    else:
        print("Column 'AQI' not found in the DataFrame.")
    return df
```

```python
## AQI class

def AQI_class(row):
    if 0 <= row['AQI'] < 64:
        return 'Good'
    elif 64 <= row['AQI'] < 170:
        return 'Moderate'
    elif 170 <= row['AQI'] < 220:
        return 'Poor'
    elif 220 <= row['AQI'] < 270:
        return 'Unhealthy'
    elif 270 <= row['AQI'] < 310:
        return 'Very Unhealthy'
    elif 310 <= row['AQI'] <= 500:
        return 'Severe'
    else:
        print(row['AQI'])
        return np.nan
```

```python
## adding a class variable to stratify

df_numeric['AQI_CLASS'] = df_numeric.apply(AQI_class, axis=1)
df_numeric.head()
```

```python
## final dataset

y = df_numeric['AQI']
class_name = df_numeric['AQI_CLASS']
X = df_numeric.drop(['AQI', 'AQI_CLASS'], axis=1)

dataset = pd.concat([X, y, class_name], axis=1)
dataset.to_csv('../data/final_dataset.csv', index=False)
```

```python
## plotting the AQI class distribution

plt.figure(figsize=(10, 6))
sns.countplot(x='AQI_CLASS', data=dataset, order=['Good', 'Moderate', 'Poor', 'Unhealthy', 'Very Unhealthy', 'Severe'])
plt.title('AQI class distribution')

plt.show()
```
